```python
def load_scan(path):
    slices = [dicom.read_file(path + '/' + s) for s in os.listdir(path)]
    slices.sort(key = lambda x: int(x.InstanceNumber))
    try:
        slice_thickness = np.abs(slices[0].ImagePositionPatient[2] - slices[1].ImagePositionPatient[2])
    except:
        slice_thickness = np.abs(slices[0].SliceLocation - slices[1].SliceLocation)
        
    for s in slices:
        s.SliceThickness = slice_thickness
        
    return slices

id=0
scans = load_scan(data_path)
imgs = np.stack([s.pixel_array for s in scans])# [::10]
# Convert to int16 (from sometimes int16), 
# should be possible as values should always be low enough (<32k)
imgs = imgs.astype(np.int16)

np.save('numpy_images', imgs)
```

```python
def sample_stack(stack, rows=5, cols=5, start_with=10, show_every=2):
    fig,ax = plt.subplots(rows,cols,figsize=[12,12])
    for i in range(rows*cols):
        ind = start_with + i*show_every
        ax[int(i/rows),int(i % rows)].set_title('slice %d' % ind)
        ax[int(i/rows),int(i % rows)].imshow(stack[ind],cmap='gray')
        ax[int(i/rows),int(i % rows)].axis('off')
    plt.show()

sample_stack(working_images)
```

```python
def plot_image(img):
    plt.figure(figsize=(6,6))
    
    plt.imshow(img, cmap=plt.cm.gray)
                
    plt.tight_layout()
    plt.show()
```

```python
def resample(image, scan, new_spacing=[2.5,2.5,2.5]):
    # Determine current pixel spacing
    spacing = map(float, ([scan[0].SliceThickness] + list(scan[0].PixelSpacing)))
    spacing = map(float, ([2.0] + list(scan[0].PixelSpacing)))


    spacing = np.array(list(spacing))
    print(scan[0].SliceThickness)

    resize_factor = spacing / new_spacing
    new_real_shape = image.shape * resize_factor
    new_shape = np.round(new_real_shape)
    real_resize_factor = new_shape / image.shape
    new_spacing = spacing / real_resize_factor
    
    image = scipy.ndimage.interpolation.zoom(image, real_resize_factor)
    
    return image, new_spacing

print ("Shape before resampling\t", working_images.shape)
imgs_after_resamp, spacing = resample(working_images, scans)
print ("Shape after resampling\t", imgs_after_resamp.shape)
```

```python
def gaus_smooth(imgs, sigma=1):

    # first build the smoothing kernel
    # sigma = 1  # width of kernel
    x = np.arange(-3, 4, 1)  # coordinate arrays -- make sure they contain 0!
    y = np.arange(-3, 4, 1)
    z = np.arange(-3, 4, 1)
    xx, yy, zz = np.meshgrid(x, y, z)
    kernel = np.exp(-(xx ** 2 + yy ** 2 + zz ** 2) / (2 * sigma ** 2))

    # apply to sample data
    data = np.zeros((11, 11, 11))
    data[5, 5, 5] = 5.
    return signal.convolve(imgs, kernel, mode="same")



smoothed_images = gaus_smooth(imgs_after_resamp)


plot_image(smoothed_images[30])
```

```python
def mophological_transforms(img, e=1, d=1):
    morphed = img
    for i in range(e):
        morphed = morphology.erosion(morphed,np.ones([3,3]))
    
    for i in range(d):
        morphed = morphology.dilation(morphed,np.ones([3,3]))
    
    return morphed

morphed_img = mophological_transforms(smoothed_images[30], e=2)
plot_image(morphed_img)
```

```python
def sheetness_filter(imgs, H, z):
    print()
    R_sheet = np.array([[0.0] * H.shape[3]] * H.shape[4])
    R_noise = np.array([[0.0] * H.shape[3]] * H.shape[4])
    R_blob = np.array([[0.0] * H.shape[3]] * H.shape[4])
    
    L1_arr = np.array([[0.0] * H.shape[3]] * H.shape[4])
    L2_arr = np.array([[0.0] * H.shape[3]] * H.shape[4])
    L3_arr = np.array([[0.0] * H.shape[3]] * H.shape[4])

    S = np.array([[0.0] * H.shape[3]] * H.shape[4])



    # get ratios at each voxel
    for x in range(H.shape[4]):
        for y in range(H.shape[3]):

            h = H[:,:,z,x,y]
            L = sorted(np.linalg.eigvals(h).real, key=np.abs)

            L3 = L3_arr[x][y] = L[0]
            L2 = L2_arr[x][y] = L[1]
            L1 = L1_arr[x][y] = L[2]
            
            if L1 != 0:
                R_sheet[x][y] = (abs(L2) / abs(L1))# ** 2
            
            if L1 != 0 and L2 != 0:
                R_blob[x][y] = abs(L3) / np.sqrt(abs(L1 * L2))
            
            R_noise[x][y] = np.sqrt((L1 ** 2) + (L2 ** 2) + (L3 ** 2))

                
        
    alpha = 0.5
    beta = 0.5 * R_noise.max()
    gamma = 0.5
    
    for x in range(H.shape[4]):
        for y in range(H.shape[3]):
            # constrain for only positive L1 to remove negative side lobe

            if L1_arr[x][y] > 0: #and L2_arr[x][y] > 0:
                a = np.e ** (-1 * (R_sheet[x][y] ** 2) / (2 * (alpha ** 2)))
                b = 1 - np.e ** (-1 * (R_noise[x][y] ** 2) / (2 * (beta ** 2)))
#                 c = 1 - np.e ** (-1 * (R_blob[x][y] ** 2) / (2 * (gamma ** 2)))

                S[x][y] = 0 if not L1_arr[x][y] else a*b#*c
            
            
#     S = 1 / (LA.norm(S)) * S
    S *= 1/S.max()  
    
    
    img = np.multiply(S, imgs[z])
    img *= 1/img.max()      
    
    return img, S, R_sheet, R_noise

filtered_image, S, R_sheet, R_noise = sheetness_filter(imgs_after_resamp, H, 30)
plot_image(imgs_after_resamp[30])
plot_image(S)


# 



thresh = 0.6

S[S > thresh] = 1
S[S <= thresh] = 0
plot_image(S)

# morphed_image = mophological_transforms(filtered_image, e=2, d=8)

# plot_image(morphed_image)

# hole_fill_image = scipy.ndimage.binary_fill_holes(morphed_image).astype(float)#mophological_transforms(filtered_image, e=1, d=1)
# plot_image(hole_fill_image)




# plot_image(np.multiply(R_sheet, R_noise))
# 
# plot_image(np.multiply(morphed_image, smoothed_images[75]))



```
